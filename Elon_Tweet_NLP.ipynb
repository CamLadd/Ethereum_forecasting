```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load in tweets
twnty15 = pd.read_csv('data/elon/2015.csv')
twnty16 = pd.read_csv('data/elon/2016.csv')
twnty17 = pd.read_csv('data/elon/2017.csv')
twnty18 = pd.read_csv('data/elon/2018.csv')
twnty19 = pd.read_csv('data/elon/2019.csv')
twnty20 = pd.read_csv('data/elon/2020.csv')
twnty21 = pd.read_csv('data/elon/2021.csv')

dataframes = [twnty15, twnty16, twnty17, twnty18, twnty19, twnty20, twnty21]
for x in dataframes:
    x.index=pd.to_datetime(x['date'])
    x.drop(columns=['date'], inplace=True)
    
df = pd.concat(dataframes)
display(df.sort_index().head())
display(df.sort_index().tail())
df = df.sort_index()
```

```python


# Rename column
df.rename(columns={'tweet':'text'}, inplace=True)

# Turn every word in every document into lowercase
from nltk.corpus import stopwords

df['text'] = df['text'].str.lower()

# Remove numbers using regular expressions
df['text'] = df['text'].str.replace('\d+', '')

# Remove special characters
# You need to put the space here or you will also remove spaces
df['text'] = df['text'].str.replace('\W+', ' ')
df['text'] = df['text'].str.replace('_', '')

# Remove white spaces
df['text'] = df['text'].str.strip()

# Remove stopwords
stop = stopwords.words('english')
for i, x in enumerate(df['text']):
    words = [word for word in df.text[i].split() if word.lower() not in stop]
    new_text = " ".join(words)
    df['text'][i] = new_text
```

```python
# Create a full word count

words = []
for i, x in enumerate(df.text):
    for t in df.text[i].split():
        words.append(t)
        
import itertools, collections

all_words = list(itertools.chain(words))
counts = collections.Counter(all_words)

'ethereum' in list(counts.keys())

counts['ethereum'], counts['eth'], counts['bitcoin'], counts['doge'], counts['btc']
```
